```python
import numpy as np
import pandas as pd
```

```python
file_to_load = "beers_df.csv"
beer_data = pd.read_csv(file_to_load)
beer_data.head(20)
```

```python
beers_df = beer_data[["brewery_name","style"]]
beers_df.head()
```

```python
 beer=beers_df.groupby(['brewery_name']).count()
# beers= beer.sort_values(['style'], ascending=False)
beer.head()
```

```python
beer_df = beers.iloc[0:30]
beer_df.head()
```

```python
data_df = beer_data[['brewery_name','rating']]
data_df= data_df.drop_duplicates()
data_df.head()
```

```python
df= pd.merge(beer,data_df, on='brewery_name', how='outer')
df.head
```

```python
df=df.sort_values(['style','rating'], ascending=False)
df
```

```python
data= df.iloc[0:30]
data
```

```python
data.to_csv('clouddata.csv', index=False)
```
